```python
from langchain_google_genai import GoogleGenerativeAI
from dotenv import load_dotenv
import os

# --- 환경 설정 ---
load_dotenv()
GOOGLE_API_KEY = os.getenv('GOOGLE_API_KEY')

if not GOOGLE_API_KEY:
    raise ValueError("환경 변수 'GOOGLE_API_KEY'를 설정해주세요. (.env 파일 확인)")
model = GoogleGenerativeAI(model="gemini-2.5-flash")
model.invoke("최근 로제가 발표한 신곡 알려줘")
```

```python
from langchain_community.tools import DuckDuckGoSearchResults
search=DuckDuckGoSearchResults(region="kr-kr", results_separator=';\n')
docs = search.invoke("도널드 트럼프 APEC")

print(docs)
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
question_answering_prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "아래 context에 기반하여 사용자의 질문에 답변하라.:\n\n{context}"
        ),
        MessagesPlaceholder(variable_name="messages"),
    ]
)

document_chain = question_answering_prompt | model
```

```python
from langchain_core.chat_history import InMemoryChatMessageHistory #langchain.memory가 아니라 langchain_core.chat_history import InMemoryChatMessageHistory로 변경
chat_history = InMemoryChatMessageHistory()
chat_history.add_user_message("도널드 프럼프?")
answer = document_chain.invoke(
    {
        "messages": chat_history.messages,
        "context": docs
    }
)

chat_history.add_ai_message(answer)

print(answer)
```

```python
links =[]

for doc in docs.split(";\n"):
    print(doc)
    link = doc.split("link:")[1].strip()
    links.append(link)
print(links)
```

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader(
    web_paths=links,
    bs_get_text_kwargs={
        "strip":True
    },
)

page_contents = []

async for doc in loader.alazy_load(): # 해당 라이브러리가 코드에서 실제로 처음 사용되는 순간 실행. 지연 로딩
    page_contents.append(doc)

for content in page_contents:
    print(content)
    print('--------------------')
```

```python
import requests
from bs4 import BeautifulSoup

def get_article_text(url):
    try:
        response = requests.get(url)
        response.raise_for_status()

        soup = BeautifulSoup(response.content, 'html.parser')
        article = soup.find('article', class_='story-news article')
        
        if article:
            return article.get_text(strip=True)
        
        else:
            try:
                if soup.find('article'):
                    return soup.find('article').get_text(strip=True)
                elif soup.find('div', id="CmAdContent"):
                    return soup.find('div', id="CmAdContent").get_text(strip=True)
            except:
                return "기사 내용을 찾을 수 없습니다."
    except requests.exceptions.RequestException as e:
        return f"URL을 가져오는 중 오류 발생: {e}"
```

```python
articles=[]
for link in links:
    print(f"URL: {link}\n")
    article_text = get_article_text(link)
    print(f"Content:\n{article_text}")
    print("-----------------------------------------")
    articles.append(article_text)
```

```python
import os
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
# (추가) Tavily 클라이언트 임포트
from tavily import TavilyClient 

# --- 환경 설정 ---
load_dotenv()
GOOGLE_API_KEY = os.getenv('GOOGLE_API_KEY')
TAVILY_API_KEY = os.getenv('TAVILY_API_KEY')
# -----------------

# (변경) LLM을 Gemini로 초기화
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", google_api_key=GOOGLE_API_KEY)

# (추가) client 객체 초기화
client = TavilyClient(api_key=TAVILY_API_KEY)

# --- (이하 동일) Step 1. 질문에 대한 자료를 TavilyClient로 가져오기 ---
query = "현대자동차 미국 시장 2025년 전망"

content = client.search(
    query, 
    include_raw_content=True,
    search_depth="advanced"
)["results"]
```

```python
prompt =[{
    "role":"system",
    "content": (
        "당신은 신문 기사를 쓰는 기자 AI입니다. \n"
        "당신은 주어진 정보를 바탕으로 객관적이고 체계적으로 작성된 기사를 써야 합니다."
        "기사의 제목은 자극적이어서 독자의 흥미를 끌 수 있도록 지어주세요."
    )
    },{
        "role" : "user",
        "content":(
            f'정보: """{content}"""\n\n'
            f'위의 정보를 사용하여, 다음 질문에 대해 자세한 보고서를 한국어로 작성하세요.: "{query}"\n'
            '-신문 기사 형식을 사용하되, MLA 표준을 준수하는 markdown 문법을 사용해주세요.'
            '-활용한 자료는 출처를 명시하세요.'
        )
    }
]
```
