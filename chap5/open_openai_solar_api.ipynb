```python
import google.generativeai as genai
from dotenv import load_dotenv
import os

# .env 파일에서 환경 변수 로드
load_dotenv()
api_key = os.getenv("GOOGLE_API_KEY")

if not api_key:
    raise ValueError("GOOGLE_API_KEY를 찾을 수 없습니다. .env 파일을 확인하세요.")

# Gemini API 키 설정
genai.configure(api_key=api_key)

# 오디오 파일 경로 설정 (Windows 경로의 `\`를 `/` 또는 `\\`로 변경)
audio_file_path = 'C:/Users/ziclp/OneDrive/Desktop/LLM Assignment/1week/PRACTICE1/chap5/G-DRAGON - POWER.mp3'

try:
    # Gemini API에 파일을 업로드합니다.
    print(f"Uploading file: {audio_file_path}")
    audio_file = genai.upload_file(path=audio_file_path)
    print(f"Successfully uploaded '{audio_file.display_name}'")

    # 오디오 파일을 처리할 모델을 선택합니다.
    model = genai.GenerativeModel(model_name="gemini-2.5-pro")

    # 프롬프트와 함께 오디오 파일을 전달하여 텍스트 변환을 요청합니다.
    print("Requesting transcription from Gemini...")
    response = model.generate_content(["이 오디오 파일을 듣고 모든 내용을 한글로 받아적어줘:", audio_file])

    # 변환된 텍스트 출력
    print("\n--- Transcription Result ---")
    print(response.text)

except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```
