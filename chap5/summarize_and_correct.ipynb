```python
import os
from typing import List, Optional
from dotenv import load_dotenv
import google.generativeai as genai
import pandas as pd
import logging

# 로깅 설정
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

class GeminiSummarizer:
    """Gemini를 사용한 회의록 요약 클래스"""
    
    def __init__(self, model_name: str = "gemini-2.5-pro"):
        """
        초기화
        Args:
            model_name: 사용할 Gemini 모델명
        """
        self.model_name = model_name
        self._load_api_key()
        
    def _load_api_key(self):
        """API 키 로드 및 SDK 설정"""
        load_dotenv()
        api_key = os.getenv("GEMINI_API_KEY") or os.getenv("GOOGLE_API_KEY")
        if not api_key:
            raise ValueError(
                "GEMINI_API_KEY 또는 GOOGLE_API_KEY가 필요합니다.\n"
                ".env 파일에 다음과 같이 설정하세요:\n"
                "GEMINI_API_KEY=your_api_key_here"
            )
        genai.configure(api_key=api_key)
        logger.info("Gemini API 키 설정 완료")
    
    def _safe_text_from_response(self, resp) -> str:
        """응답에서 안전하게 텍스트 추출"""
        if not getattr(resp, "candidates", None):
            return "[No candidates returned]"
        
        cand = resp.candidates[0]
        finish_reason = getattr(cand, "finish_reason", None)
        
        # 정상적인 텍스트 추출 시도
        content = getattr(cand, "content", None)
        if content and getattr(content, "parts", None):
            text_parts = []
            for part in content.parts:
                if getattr(part, "text", None):
                    text_parts.append(part.text)
            
            if text_parts:
                return "\n".join(text_parts).strip()
        
        # 텍스트가 없는 경우 원인 분석
        reason_map = {
            1: "STOP (정상 완료)",
            2: "MAX_TOKENS (토큰 한계)",
            3: "SAFETY (안전 필터)",
            4: "RECITATION (저작권 우려)",
            5: "OTHER (기타 오류)"
        }
        
        reason_str = reason_map.get(finish_reason, f"UNKNOWN({finish_reason})")
        
        # 안전 등급 정보 추가
        safety_info = ""
        safety_ratings = getattr(cand, "safety_ratings", None)
        if safety_ratings:
            safety_details = []
            for rating in safety_ratings:
                category = getattr(rating, "category", "UNKNOWN")
                probability = getattr(rating, "probability", "UNKNOWN")
                safety_details.append(f"{category}:{probability}")
            safety_info = " | Safety: " + ", ".join(safety_details)
        
        return f"[Empty content] Finish reason: {reason_str}{safety_info}"
    
    def _validate_text(self, text: str) -> bool:
        """텍스트 유효성 검사"""
        if not text or not text.strip():
            logger.error("입력 텍스트가 비어있습니다.")
            return False
        
        if len(text.strip()) < 10:
            logger.warning("입력 텍스트가 너무 짧습니다 (10자 미만)")
            return False
        
        return True
    
    def _chunk_text(self, text: str, chunk_chars: int = 6000, overlap: int = 300) -> List[str]:
        """텍스트를 청크로 분할 (보수적인 크기 설정)"""
        text = text.strip()
        if len(text) <= chunk_chars:
            return [text]
        
        chunks = []
        sentences = text.split('. ')  # 문장 단위로 우선 분할 시도
        
        current_chunk = ""
        for sentence in sentences:
            # 현재 청크에 문장을 추가했을 때의 길이 확인
            potential_length = len(current_chunk) + len(sentence) + 2  # ". " 추가
            
            if potential_length <= chunk_chars:
                current_chunk += sentence + ". "
            else:
                # 현재 청크가 비어있지 않으면 저장
                if current_chunk.strip():
                    chunks.append(current_chunk.strip())
                
                # 새 청크 시작
                current_chunk = sentence + ". "
        
        # 마지막 청크 추가
        if current_chunk.strip():
            chunks.append(current_chunk.strip())
        
        logger.info(f"텍스트를 {len(chunks)}개 청크로 분할")
        return chunks
    
    def _summarize_once(
        self, 
        system_prompt: str, 
        content: str, 
        max_output_tokens: int = 2048,
        temperature: float = 0.3
    ) -> str:
        """단일 요약 수행"""
        if not self._validate_text(content):
            return "[Invalid input text]"
        
        try:
            model = genai.GenerativeModel(
                model_name=self.model_name,
                system_instruction=system_prompt,
            )
            
            logger.debug(f"입력 텍스트 길이: {len(content)} 문자")
            
            resp = model.generate_content(
                content,
                generation_config={
                    "temperature": temperature,
                    "max_output_tokens": max_output_tokens,
                    "top_p": 0.8,
                    "top_k": 40,
                },
                request_options={"timeout": 300},  # 타임아웃 증가
            )
            
            result = self._safe_text_from_response(resp)
            logger.debug(f"출력 텍스트 길이: {len(result)} 문자")
            
            return result
            
        except Exception as e:
            logger.error(f"요약 생성 중 오류 발생: {str(e)}")
            return f"[Error] {str(e)}"
    
    def summarize_meeting(
        self, 
        transcript: str, 
        system_prompt: Optional[str] = None
    ) -> str:
        """
        회의록 요약 (Map-Reduce 방식)
        
        Args:
            transcript: 회의록 텍스트
            system_prompt: 시스템 프롬프트 (선택사항)
        
        Returns:
            str: 요약된 회의록
        """
        if not self._validate_text(transcript):
            return "[Error] 유효하지 않은 입력 텍스트"
        
        # 기본 시스템 프롬프트
        if not system_prompt:
            system_prompt = """
당신은 회의 내용을 분석하고 요약하는 전문가입니다. 다음 지침을 따라 요약해주세요:

**요약 형식 (마크다운):**
# 회의 요약

## 🎯 핵심 결론
- 주요 결정사항들을 간결하게

## 📋 주요 논의사항
- 중요한 토론 내용과 의견들

## ⚡ 액션 아이템
- [ ] 구체적 작업 (담당자/기한)
- [ ] 후속 조치사항

## 💡 기타 특이사항
- 추가로 언급할 만한 내용

**주의사항:**
- 원문에 없는 내용을 추가하지 마세요
- 객관적이고 정확한 정보만 포함하세요
- 중복된 내용은 제거하세요
"""

        logger.info("회의록 요약 시작")
        logger.info(f"입력 텍스트 길이: {len(transcript):,} 문자")
        
        # 1. 청킹
        chunks = self._chunk_text(transcript, chunk_chars=5000, overlap=200)
        
        if len(chunks) == 1:
            # 짧은 텍스트는 바로 요약
            logger.info("단일 청크 요약 수행")
            return self._summarize_once(system_prompt, chunks[0], max_output_tokens=2048)
        
        # 2. 부분 요약들 생성 (Map 단계)
        logger.info(f"Map 단계: {len(chunks)}개 청크 개별 요약")
        partial_summaries = []
        
        for idx, chunk in enumerate(chunks, 1):
            logger.info(f"부분 요약 {idx}/{len(chunks)} 진행 중...")
            
            content = f"""
다음은 회의록의 일부분입니다 ({idx}/{len(chunks)}). 
이 부분의 핵심 내용을 요약해주세요. 주요 논의사항, 결정사항, 액션아이템을 중심으로 정리하되, 
최종 통합 요약에서 사용할 수 있도록 간결하면서도 완전한 정보를 포함해주세요.

---
{chunk}
"""
            
            partial_summary = self._summarize_once(
                system_prompt="회의록 부분 요약을 수행하는 전문가입니다. 핵심 내용을 놓치지 않고 간결하게 요약합니다.",
                content=content,
                max_output_tokens=1024
            )
            
            # 오류 체크
            if partial_summary.startswith("["):
                logger.warning(f"부분 요약 {idx} 생성 실패: {partial_summary}")
                partial_summaries.append(f"[부분 {idx}] 요약 생성 실패")
            else:
                partial_summaries.append(f"**[부분 {idx}/{len(chunks)}]**\n{partial_summary}")
        
        # 3. 통합 요약 (Reduce 단계)
        logger.info("Reduce 단계: 부분 요약들을 통합")
        
        merged_content = "\n\n---\n\n".join(partial_summaries)
        
        final_prompt = f"""
다음은 긴 회의록을 여러 부분으로 나누어 요약한 결과들입니다. 
이들을 통합하여 완전한 하나의 회의 요약본을 만들어주세요.

중복된 내용은 제거하고, 일관성 있는 구조로 정리해주세요.
앞서 제시한 마크다운 형식을 엄격히 따라주세요.

---
{merged_content}
"""
        
        final_summary = self._summarize_once(
            system_prompt=system_prompt,
            content=final_prompt,
            max_output_tokens=3000
        )
        
        logger.info("회의록 요약 완료")
        return final_summary

def process_meeting_transcript(df_final: pd.DataFrame) -> str:
    """
    STT+화자분리 결과를 회의록으로 변환 후 요약
    
    Args:
        df_final: STT+화자분리 최종 결과 DataFrame
    
    Returns:
        str: 요약된 회의록
    """
    try:
        # 데이터 검증
        if df_final.empty:
            logger.error("입력 DataFrame이 비어있습니다.")
            return "[Error] 빈 데이터프레임"
        
        if 'text' not in df_final.columns:
            logger.error("'text' 컬럼이 DataFrame에 없습니다.")
            return "[Error] 'text' 컬럼 없음"
        
        # 텍스트 결합 (화자 정보와 함께)
        transcript_parts = []
        
        for _, row in df_final.iterrows():
            speaker = row.get('speaker_id', 'Unknown')
            text = row.get('text', '').strip()
            start_time = row.get('start', 0)
            
            if text:
                # 시간 정보를 분:초 형식으로 변환
                minutes = int(start_time // 60)
                seconds = int(start_time % 60)
                time_str = f"{minutes:02d}:{seconds:02d}"
                
                transcript_parts.append(f"[{time_str}] {speaker}: {text}")
        
        if not transcript_parts:
            logger.error("유효한 텍스트가 없습니다.")
            return "[Error] 유효한 텍스트 없음"
        
        transcript = "\n\n".join(transcript_parts)
        
        logger.info(f"회의록 텍스트 생성 완료: {len(transcript_parts)}개 발언, {len(transcript):,} 문자")
        
        # 요약 수행
        summarizer = GeminiSummarizer()
        summary = summarizer.summarize_meeting(transcript)
        
        return summary
        
    except Exception as e:
        logger.error(f"회의록 처리 중 오류 발생: {str(e)}")
        return f"[Error] {str(e)}"

# 사용 예시
if __name__ == "__main__":
    # 예시: STT+화자분리 결과가 있다고 가정
    df_final = pd.read_csv(r"C:\Users\전산2실 PC\Downloads\PRACTICE1\chap5\output\싼기타_비싼기타_final.csv", sep="|")
    summary = process_meeting_transcript(df_final)
    print(summary)

with open(r"C:\Users\전산2실 PC\Downloads\PRACTICE1\chap5\output\meeting_summary.md", "w", encoding='utf-8') as f:
    f.write(summary)
```

```python
# 1. 라이브러리 설치
#%pip install google-generativeai python-dotenv pandas

import google.generativeai as genai
import os
import pandas as pd
from dotenv import load_dotenv

# .env 파일에서 환경 변수 로드
load_dotenv()

# 2. API 키 설정
# .env 파일에 GOOGLE_API_KEY="YOUR_API_KEY" 형식으로 키를 저장하세요.
GOOGLE_API_KEY = os.getenv('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

# 3. Gemini 모델 설정
model = genai.GenerativeModel('gemini-2.5-flash') # 또는 다른 Gemini 모델

summary = "싼 기타와 비싼 기타 중 어떤 것으로 시작할지에 대한 토론"

# 1. df_rttm에서 필요한 'name', 'text' 컬럼만 복사하여 df_meeting_note를 만듭니다.
df_meeting_note = df_rttm[['name', 'text']].copy()

# 2. df_meeting_note를 API가 처리하기 좋은 리스트-딕셔너리 형태로 변환합니다.
meeting_note_dict = df_meeting_note.to_dict('records')
# --- 데이터 준비 끝 ---


# --- STT 결과 보정 루프 ---
for row in meeting_note_dict:
    if not isinstance(row['text'], str):
        continue

    # Gemini API에 맞는 프롬프트 형식으로 수정
    correction_prompt = f'''
        너는 주어진 회의 녹취록을 수정 및 보완하는 봇이다.
        주어진 회의 녹취록은 STT로 작성된 결과이므로, 이 중에 오류가 있는 부분을 찾아내고 수정하라.

        표현, 말투는 최대한 원본과 일치하도록 유지하되, 잘못 받아적은 내용을 수정하라.
        원본의 내용을 최대한 살리되, 잘못된 내용만 수정하여 수정본을 작성하여라
        어떤 경우에는 A라는 사람이 말한 내용을 B가 말한 것으로 잘못 기록된 경우도 있을 수 있다. 이런 경우에는 문맥을 고려하여 수정하라.

        원본 내용 중 빠지거나, 추가되거나, 잘못 기록된 부분을 찾아내어 수정하라. 수정된 메시지 이외에는 아무것도 작성하지 말라.
        --------------------------------
        회의 요약문 : {summary}
        --------------------------------
        회의 녹취록 전문: {df_meeting_note.to_json(orient='records', force_ascii=False)}
        --------------------------------
        확인할 메시지 원본: {row['text']}
        --------------------------------
        수정된 메시지:
    '''

    # Gemini API 호출
    response = model.generate_content(
        correction_prompt,
        generation_config=genai.types.GenerationConfig(temperature=0.1)
    )

    correction = response.text
    if ':' in correction:  # “수정: “ 과 같은 형식으로 반환되는 경우 해결
        correction = correction.split(':')[-1]

    correction = correction.strip() # 좌우 공백 제거

    print(f"이름: {row['name']}")
    print(f"원본: {row['text']}")
    print(f"수정: {correction}")
    print('-----------------')

    row['corrected_text'] = correction
```

```python
# 1. 라이브러리 설치
%pip install openai python-dotenv pandas

from openai import OpenAI
import os
import pandas as pd
from dotenv import load_dotenv

# .env 파일에서 환경 변수 로드
load_dotenv()

# 2. API 키 설정 및 클라이언트 초기화
# .env 파일에 UPSTAGE_API_KEY="YOUR_API_KEY" 형식으로 키를 저장하세요.
UPSTAGE_API_KEY = os.getenv('UPSTAGE_API_KEY')
client = OpenAI(
    api_key=UPSTAGE_API_KEY,
    base_url="https://api.upstage.ai/v1/solar"
)

summary = "싼 기타와 비싼 기타 중 어떤 것으로 시작할지에 대한 토론"

# 1. df_rttm에서 필요한 'name', 'text' 컬럼만 복사하여 df_meeting_note를 만듭니다.
df_meeting_note = df_rttm[['name', 'text']].copy()

# 2. df_meeting_note를 API가 처리하기 좋은 리스트-딕셔너리 형태로 변환합니다.
meeting_note_dict = df_meeting_note.to_dict('records') # records는 데이터프레임의 각 행을 하나의 딕셔너리로 만들고, 이 딕셔너리들을 리스트에 담아 반환하라는 의미
# --- 데이터 준비 끝 ---


# --- STT 결과 보정 루프 ---
for row in meeting_note_dict:
    if not isinstance(row['text'], str):
        continue

    # Upstage API에 맞는 프롬프트 형식 (Gemini와 동일하게 사용 가능)
    correction_system_prompt = f'''
        너는 주어진 회의 녹취록을 수정 및 보완하는 봇이다.
        주어진 회의 녹취록은 STT로 작성된 결과이므로, 이 중에 오류가 있는 부분을 찾아내고 수정하라.

        표현, 말투는 최대한 원본과 일치하도록 유지하되, 잘못 받아적은 내용을 수정하라.
        원본의 내용을 최대한 살리되, 잘못된 내용만 수정본을 작성하여라.
        수정의도나 수정한 까닭은 적지 않는다.
        마크다운 문자는 쓰지 않는다.
        

        어떤 경우에는 A라는 사람이 말한 내용을 B가 말한 것으로 잘못 기록된 경우도 있을 수 있다. 이런 경우에는 문맥을 고려하여 수정하라.

        원본 내용 중 빠지거나, 추가되거나, 잘못 기록된 부분을 찾아내어 수정하라. 수정된 메시지 이외에는 아무것도 작성하지 말라.
        --------------------------------
        회의 요약문 : {summary}
        --------------------------------
        회의 녹취록 전문: {df_meeting_note.to_json(orient='records', force_ascii=False)}
        --------------------------------
        확인할 메시지 원본: {row['text']}
    '''

    # Upstage API 호출
    response = client.chat.completions.create(
        model='solar-pro2',  # Upstage 모델명으로 변경
        temperature=0.1,
        messages=[
            {"role": "system", "content": correction_system_prompt},
        ]
    )

    correction = response.choices[0].message.content
    if ':' in correction:
        correction = correction.split(':')[-1]
        
    correction = correction.strip()

    print(f"이름: {row['name']}")
    print(f"원본: {row['text']}")
    print(f"수정: {correction}")
    print('-----------------')

    row['corrected_text'] = correction
```
